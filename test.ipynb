```
#데이터 확인 및 필요 없는 열 삭제
#df=pd.read_csv(io.BytesIO(uploaded['bundang1.csv']),encoding='euc-kr')
#df=pd.read_csv("bundang1.csv")
pd.set_option('display.max_columns', 500)

df=pd.read_csv("bundang3years2.csv")
print(df.head())

df.drop(['시군구'],axis=1,inplace=True)
df.drop(['단지명'],axis=1,inplace=True)
df.drop(['건축년도'],axis=1,inplace=True)
# df.drop(['마트최소거리'],axis=1,inplace=True)
# df.drop(['지하철도보거리'],axis=1,inplace=True)
# df.drop(['용적률'],axis=1,inplace=True)
# df.drop(['대형병원최소거리'],axis=1,inplace=True)
#df.drop(['층'],axis=1,inplace=True)
# df.drop(['건폐율'],axis=1,inplace=True)
#df.drop(['총세대수'],axis=1,inplace=True)
# df.drop(['규모별월세가격지수'],axis=1,inplace=True)
# df.drop(['규모별수익률'],axis=1,inplace=True)
# df.drop(['동별인구수'],axis=1,inplace=True)
print(df.head())

#데이터 형태 정수로 변환 
Y = df['거래금액'].values
Y = Y.astype('int')
#print(type(Y[0]))
X = df.drop(labels=['거래금액'],axis=1)
X = X.astype('float')

# 훈련셋, 시험셋 분류 
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)



```

```
#-----------------------------------------------최종정리
# 최종정리 3개 기법 반복시행 후 결과도출
# 정규화 한거랑 안한거랑 별 차이없음.
from sklearn.ensemble import RandomForestRegressor
from sklearn import svm
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import scipy.stats

RF_SUM_train=0
SVR_SUM_train=0
LR_SUM_train=0
RF_SUM_test=0
SVR_SUM_test=0
LR_SUM_test=0

#paired t-test list
RF_test_list =[]
SVR_test_list =[]
LR_test_list =[]

#반복횟수
k=10

for i in range(k):

  # 훈련셋, 시험셋 분류 
  X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

  print("----------",i+1,"----------")
  #랜덤포레스트
  RF_model = RandomForestRegressor(n_estimators=5, max_depth=5, max_features=9)
  RF_model.fit(X_train, Y_train)
  #예측값 생성
  RF_prediction_test = RF_model.predict(X_test)
  RF_prediction_test = RF_prediction_test.astype('int')
  #만든 모델에 train set 다시 넣어 trainset rmse 구하기 위해 만든것
  RF_train_test = RF_model.predict(X_train)
  RF_train_test = RF_train_test.astype('int')
  #훈련셋 RMSE
  RF_RMSE_train =np.sqrt(mean_squared_error(Y_train, RF_train_test))
  RF_RMSE_train = RF_RMSE_train.astype('int')
  print('RF_RMSE_train:',RF_RMSE_train)
  #시험셋 RMSE
  RF_RMSE =np.sqrt(mean_squared_error(Y_test, RF_prediction_test))
  RF_RMSE = RF_RMSE.astype('int')
  print('RF_RMSE_test:',RF_RMSE)
  

  #MIN-MAX Scaler
  minmax_scaler = MinMaxScaler()
  minmax_scaler.fit(X_train)
  X_train_scaled = minmax_scaler.transform(X_train)
  X_test_scaled = minmax_scaler.transform(X_test)

  #SVR
  SVR_model = svm.SVR(kernel='rbf',gamma=0.01,C=10000)
  SVR_model.fit(X_train, Y_train)
  #예측값 생성
  SVR_prediction_test = SVR_model.predict(X_test)
  SVR_prediction_test = SVR_prediction_test.astype('int')
  #만든 모델에 train set 다시 넣어 trainset rmse 구하기 위해 만든것
  SVR_train_test = SVR_model.predict(X_train)
  SVR_train_test = SVR_train_test.astype('int')
  #훈련셋 RMSE
  SVR_RMSE_train =np.sqrt(mean_squared_error(Y_train, SVR_train_test))
  SVR_RMSE_train = SVR_RMSE_train.astype('int')
  print('SVR_RMSE_train:',SVR_RMSE_train)
  #시험셋 RMSE
  SVR_RMSE =np.sqrt(mean_squared_error(Y_test, SVR_prediction_test))
  SVR_RMSE = SVR_RMSE.astype('int')
  print('SVR_RMSE_test:',SVR_RMSE)

  #다중회귀
  LR_model=LinearRegression()
  LR_model.fit(X_train, Y_train)
  LR_prediction_test = LR_model.predict(X_test)
  LR_prediction_test = LR_prediction_test.astype('int')
  LR_train_test = LR_model.predict(X_train)
  LR_train_test = LR_train_test.astype('int')
  #훈련셋 RMSE
  LR_RMSE_train =np.sqrt(mean_squared_error(Y_train, LR_train_test))
  LR_RMSE_train =LR_RMSE_train.astype('int') 
  print('LR_RMSE_train:',LR_RMSE_train)
  #시험셋 RMSE
  LR_RMSE =np.sqrt(mean_squared_error(Y_test, LR_prediction_test))
  LR_RMSE =LR_RMSE.astype('int') 
  print('LR_RMSE_test:',LR_RMSE)

  RF_SUM_train += RF_RMSE_train
  SVR_SUM_train += SVR_RMSE_train
  LR_SUM_train += LR_RMSE_train
  RF_SUM_test += RF_RMSE
  SVR_SUM_test += SVR_RMSE
  LR_SUM_test += LR_RMSE

  RF_test_list.append(RF_RMSE)
  SVR_test_list.append(SVR_RMSE)
  LR_test_list.append(LR_RMSE)
  

print("----------total ",k," loop summary----------")
print("RF_RMSE_MEAN_train: ",RF_SUM_train/k)
print("RF_RMSE_MEAN_test: ",RF_SUM_test/k)
print("SVR_RMSE_MEAN_train: ",SVR_SUM_train/k)
print("SVR_RMSE_MEAN_test: ",SVR_SUM_test/k)
print("LR_RMSE_MEAN_train: ",LR_SUM_train/k)
print("LR_RMSE_MEAN_test: ",LR_SUM_test/k)

print("----------paired t-test list----------")
print("RF_test_list: ",RF_test_list)
print("SVR_test_list: ",SVR_test_list)
print("LR_test_list: ",LR_test_list)

print("----------paired t-test result----------")
print("----------RF vs SVR----------")
RF_SVR_ttest=scipy.stats.ttest_rel(RF_test_list, SVR_test_list)
print(RF_SVR_ttest)
print("----------SVR vs LR----------")
SVR_LR_ttest=scipy.stats.ttest_rel(SVR_test_list, LR_test_list)
print(SVR_LR_ttest)
print("----------LR vs RF----------")
LR_RF_ttest=scipy.stats.ttest_rel(LR_test_list, RF_test_list)
print(LR_RF_ttest)


```
